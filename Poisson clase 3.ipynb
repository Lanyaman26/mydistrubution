```python
from scipy.stats import poisson 
  
import numpy as np 
  
import matplotlib.pyplot as plt 
  
  
x = np.arange(0, 100, 0.5) 
  
y = poisson.pmf(x, mu=40, loc=10) 
  
  
plt.plot(x, y) 
  
plt.show() 
```

```python
#El sitio web de una tienda online de ropa recibe una media de 0.1 ordenes de compra por segundo.
# Suponiendo que las peticiones al sitio web son independientes y a ritmo constante, 
#¿cual es la probabilidad de un maximo de 5 peticiones en un segundo?
#si el sistema de cobro de la tienda es deficiente cuando recibe mas de 6 ordenes de compra 
#¿cual es la probabilidad de que ocurra esto?
import scipy.stats as ss
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
import seaborn as sns
```

```python
A=ss.poisson(0.1)
b=A.cdf(5)
print(b)
```

```python
Y=ss.poisson(6)#numero de ordenes de compra o x
c=Y.sf(10)
print(c)
```

```python
#graficar distribucion
y=np.arange(20)
plt.plot(y,Y.pmf(y),"ro")
plt.vlines(y,0,Y.pmf(y),"y")
plt.title("Distribucion de poisson")
plt.ylabel("probabilidad")
plt.xlabel("valores")
plt.show()
```

```python
#ver histograma
sns.distplot(random.poisson(lam=5,size=200),kde=False)
plt.title("Histograma de distribucion Poisson")
plt.show()
```
